```python
from PIL import Image 
import requests
from io import BytesIO
for i in range(170, len(temp)):
    img = temp.iloc[i]
    name = img['name']
#     url1 = requests.get(img['img_final'])
    url1 = requests.get(img['img1'])
    img1 = Image.open(BytesIO(url1.content))
    print(i,name)
    display(img1)
```
